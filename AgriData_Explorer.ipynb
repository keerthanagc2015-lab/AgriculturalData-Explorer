```python
#•	Sorghum Production (Kharif and Rabi) by Region

def get_region(state):
    if state in ["Punjab", "Haryana", "Uttar Pradesh", "Rajasthan"]:
        return "North"
    elif state in ["Tamil Nadu", "Karnataka", "Andhra Pradesh", "Kerala"]:
        return "South"
    elif state in ["Maharashtra", "Gujarat", "Goa"]:
        return "West"
    elif state in ["West Bengal", "Bihar", "Odisha", "Jharkhand"]:
        return "East"
    else:
        return "Central/Other"

df["Region"] = df["State Name"].apply(get_region)


region_data = df.groupby("Region").agg({
    "KHARIF SORGHUM PRODUCTION (1000 tons)": "sum",
    "RABI SORGHUM PRODUCTION (1000 tons)": "sum"
})


plt.figure(figsize=(10,5))

region_data.plot(
    kind='bar',
    color=['#8FD9A8', '#F4A261']
)

plt.title("Sorghum Production (Kharif vs Rabi) by Region", fontsize=14, fontweight='bold')
plt.xlabel("Region")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=45)

plt.legend(["Kharif", "Rabi"])
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
print("📊 Insights: Sorghum Production (Kharif vs Rabi) by Region")

print("1. The West region leads sorghum production in both Kharif and Rabi seasons, indicating strong regional dominance.")
print("2. Kharif production is consistently higher than Rabi across most regions, reflecting dependence on monsoon conditions.")
print("3. The South region shows nearly equal production in both seasons, suggesting stable and balanced cultivation practices.")
print("4. Central/Other regions exhibit high Kharif but significantly lower Rabi production, indicating seasonal dependency.")
print("5. The North region shows moderate Kharif production with minimal Rabi contribution.")
print("6. Overall, sorghum production varies significantly across regions and seasons, driven by climatic and agricultural factors.")
```

```python
#•	Soybean Production by Top 5 States and Yield Efficiency


# Step 1: Total production by state
soy_prod = df.groupby("State Name")["SOYABEAN PRODUCTION (1000 tons)"].sum()

# Top 5 states
top5_states = soy_prod.sort_values(ascending=False).head(5)

# Step 2: Get yield for those states
soy_yield = df.groupby("State Name")["SOYABEAN YIELD (Kg per ha)"].mean()

# Filter only top 5 states
top5_yield = soy_yield[top5_states.index]

# Combine into one DataFrame
final_df = pd.DataFrame({
    "Production": top5_states,
    "Yield": top5_yield
})

print(final_df)
```
